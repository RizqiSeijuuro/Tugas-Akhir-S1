# 1. Import

```python
# Import Library

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import Data
df = pd.read_csv('https://raw.githubusercontent.com/RizqiSeijuuro/Tugas-Akhir-S1/main/Data%20Penelitian%20TA/Konsistensi.csv', sep=',')
df
```

# 2. Data Understanding

```python
# Read DataFrame
df.describe()
```

## 2.1. Data Preparation

```python
# Drop Missing Values & Change data type to datetime
df = df.dropna()
df['Date'] = pd.to_datetime(df['Date'])
pd.to_datetime(df['Time2'], format='%H:%M:%S')
```

```python
df
```

# 3. Data Visualization

```python
# Create Data Visualization
fig, ax = plt.subplots(figsize=(18, 8), dpi=216)
plt.title("Uji Konsistensi Lampu UV", fontsize=22, y=1.05)
plt.xlabel('Detik ke', fontsize='18')
plt.yticks(fontsize='14')
plt.ylabel(r'$ Intensitas \quad UV \quad (\frac{mW}{cm^2})$', fontsize='18')
ax.plot(df.index, df['Intensity'], linewidth=1.1, alpha=0.9)
```

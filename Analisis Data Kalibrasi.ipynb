```python
# Create Function to Visualise Histogram & Boxplot
def vis_univar(data, column):
    sns.color_palette("pastel")
    fig, (ax_box, ax_hist) = plt.subplots(2, sharex = True, gridspec_kw = {'height_ratios':(.2, .8)})
    
    sns.boxplot(data = data, x = column, ax = ax_box)
    sns.histplot(data = data, x = column, ax = ax_hist, bins=10)
    
    ax_box.set(xlabel = '')
    ax_hist.set(ylabel = 'Histogram (Count)')
    
    plt.show()
```

```python
vis_univar(dataLampu, 'Sensor UV')
```

```python
vis_univar(dataLampu, 'UV Meter')
```

```python
# Get Mean & Correlation from dataset
print('Kalibrasi mengukur Lampu UV')
print('-'*50)
print('Rerata intensitas sensor UV:', dataLampu['Sensor UV'].mean())
print('Rerata intensitas UV Meter:', dataLampu['UV Meter'].mean())
print('Korelasi Pearson (Lampu UV):', dataLampu['Sensor UV'].corr(dataLampu['UV Meter'], 'pearson'))
print('Korelasi Spearman (Lampu UV):', dataLampu['Sensor UV'].corr(dataLampu['UV Meter'], 'spearman'))

print('\n\nKalibrasi mengukur Matahari')
print('-'*50)
print('Rerata intensitas sensor UV:', dataMatahari['Sensor UV'].mean())
print('Rerata intensitas UV Meter:', dataMatahari['UV Meter'].mean())
print('Korelasi Pearson (Lampu UV):', dataMatahari['Sensor UV'].corr(dataMatahari['UV Meter'], 'pearson'))
print('Korelasi Spearman (Lampu UV):', dataMatahari['Sensor UV'].corr(dataMatahari['UV Meter'], 'spearman'))
```

# 3. Data Visualization

```python
# Create Visualization (line plot)

x = dataLampu['Waktu']
y1 = dataLampu['Sensor UV']
y2 = dataLampu['UV Meter']

fig = plt.figure(figsize = (18,8), dpi=216)
ax = plt.axes()

line1, = plt.plot(x, y1, color = 'Indigo', label = 'Sensor UV', marker="X", markersize=10, markerfacecolor="red", markeredgecolor="black")
line2, = plt.plot(x, y2, color = 'darkgreen', label = 'UV Meter', marker="o", markersize=15, markerfacecolor="None")

plt.legend(handles = [line1, line2], loc = 'best')

plt.grid(color = 'blue', which = 'major', alpha = 0.2)
plt.minorticks_on()
plt.grid(color = 'pink', which = 'minor', alpha = 0.2)
plt.xticks(rotation = 50, fontsize='14')
plt.yticks(fontsize='14')

ax.set_xlabel('Waktu', fontsize='18')
ax.set_ylabel(r'$ Intensitas \quad UV \quad (\frac{mW}{cm^2})$', fontsize='18')
ax.xaxis.set_label_coords(.5, -.13)
ax.yaxis.set_label_coords(-.03, .5)
ax.set_ylim(0, 10)
ax.set_title('Pengukuran UV menggunakan Sensor UV dan UV Meter (Lampu UV)', fontsize='22', y=1.05)

plt.show()
```

# 4. Evaluation Metric

### Mean Absolute Percent Formula
MAPE Formula

```python
# Create Function to get Mean Absolute Percentage Error (MAPE)
def MAPE(y,y_pred):
    return np.mean(np.abs((y-y_pred)/y))*100
```

```python
print('MAPE for Lampu UV:', MAPE(dataLampu['Sensor UV'], dataLampu['UV Meter']),'%')
print('MAPE for Matahari:', MAPE(dataMatahari['Sensor UV'], dataMatahari['UV Meter']),'%')
```

# 1. Import

```python
# Import Library

from PIL import Image
import urllib.request
import statistics as s
import pandas as pd
```

# 2. RGB Analysis

```python
# Create Function to get RGB Value from Images

def getRGBInfo(path):
  R_Values = []
  G_Values = []
  B_Values = []
  im = Image.open(path, 'r')
  width, height = im.size
  pixel_values = list(im.getdata())
  for i in pixel_values:
    R_Values.append(i[0])
    G_Values.append(i[1])
    B_Values.append(i[2])
  return pd.DataFrame({'Nilai' : ['Rerata', 'Min', 'Max'],
                       'R' : [s.mean(R_Values), min(R_Values), max(R_Values)],
                       'G' :[ s.mean(G_Values), min(G_Values), max(G_Values)],
                       'B' : [s.mean(B_Values), min(B_Values), max(B_Values)]})
```

```python
# Get RGB VALUE from Tempe before treatment

urllib.request.urlretrieve(
  'https://raw.githubusercontent.com/RizqiSeijuuro/Tugas-Akhir-S1/main/Data%20Penelitian%20TA/Efek%20UVC/Tempe_Sebelum.jpg',
   "Tempe_Sebelum.jpg")
tempe_sebelum = getRGBInfo("Tempe_Sebelum.jpg")
tempe_sebelum
```

```python
# Get RGB VALUE from Tempe after treatment

urllib.request.urlretrieve(
  'https://raw.githubusercontent.com/RizqiSeijuuro/Tugas-Akhir-S1/main/Data%20Penelitian%20TA/Efek%20UVC/Tempe_Sesudah.jpg',
   "Tempe_Sesudah.jpg")
tempe_sesudah = getRGBInfo("Tempe_Sesudah.jpg")
tempe_sesudah
```
